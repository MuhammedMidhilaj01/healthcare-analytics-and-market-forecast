### duplicates

```python
df = df.drop_duplicates()
```

```python
df.shape
```

### Convert Age to Years

```python
df["age_years"] = (df["age"] / 365).round(0)
```

### Create BMI Column

```python
df["BMI"] = (df["weight"] / ((df["height"]/100)**2)).round(1)
```

### Create Age Groups

```python
bins = [0,40,50,60,100]
labels = ["<40","40-50","50-60","60+"]

df["age_group"] = pd.cut(df["age_years"], bins=bins, labels=labels)
```

### Check Blood Pressure Outliers

```python
df[(df["ap_hi"] < 80) | (df["ap_hi"] > 250)]
df[(df["ap_lo"] < 40) | (df["ap_lo"] > 150)]
```

```python
df = df[(df["ap_hi"] < 250) & (df["ap_hi"] > 80)]
df = df[(df["ap_lo"] < 150) & (df["ap_lo"] > 40)]
```
